```python
chessboard = np.zeros((8,8))
print(chessboard)
```

```python
# chessboard[0][1] = 1
# chessboard[1][0] = 1

# Numpy Array
chessboard[1::2, ::2] = 1
chessboard[::2, 1::2] = 1

print(chessboard)
```

```python
# Algorithmic Assignment: Plot Chessboard
chessboard = np.zeros((8,8), dtype=int)

chessboard[1::2, ::2] = 1
chessboard[::2, 1::2] = 1

# Python List
chessboard = chessboard.tolist()

black = '\u2B1B'
white = '\u2B1C'
white_king = '\u2655'

chessboard[0][0] = black
chessboard[0][1] = white
chessboard[0][4] = white_king

# print(chessboard)
# print(type(chessboard))
for row in chessboard:
    print(row)
```

```python
scores = np.array([
                    [80, 85, 90],
                    [70, 75, 80],
                    [90, 95, 100],    
                    [60, 65, 70]        
                  ])
print(scores)
print(scores.shape)
```

```python
another_scores = np.array([
                    [80, 85, 90],
                    [60, 65, 70]        
                  ])

print(another_scores)
print(another_scores.shape)
```

```python
more_scores = np.array([
                    [80],
                    [60],
                    [115],
                    [97]
                  ])

print(more_scores)
print(more_scores.shape)
```

```python
# Stacking Arrays -> Concatenating Arrays
result1 = np.vstack((scores, another_scores))
print(result1)
print(result1.shape)
```

```python
result2 = np.hstack((scores, more_scores))
print(result2)
print(result2.shape)
```
