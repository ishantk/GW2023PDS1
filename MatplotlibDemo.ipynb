```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
print(type(df))
```

```python
id_column = df['id']
print(id_column)
print(type(id_column))
```

```python
# Line Chart
X = df['season']
Y = df['win_by_runs']

# Figure Function is used to mention the size of graph
plt.figure(figsize=(10, 10))
# plt.plot(X, Y)
# plt.bar(X, Y)
# plt.barh(X, Y)
plt.scatter(X, Y)
plt.xlabel('season')
plt.ylabel('win by runs')
plt.title('Win By Runs in IPL Seasons')
plt.savefig('IPL-WINS-SEASON.png')
plt.show()
```

```python
# Histogram
X = df['season']
Y = df['win_by_runs']
plt.hist(Y, color='green', bins=10, edgecolor='red')
plt.xlabel('win_by_runs')
plt.ylabel('Frequency')
plt.legend(['wins'])
plt.grid(True)
plt.title('Win By Runs in IPL Seasons - Histogram', fontsize=14, fontstyle='italic', fontweight='bold')
plt.show()
```

```python
# Pie Chart
labels = df['winner'].unique()
print(labels)
print(len(labels))
```

```python
sizes = df['winner'].value_counts()
print(sizes)
```

```python
plt.figure(figsize=(12, 12))
plt.pie(sizes, autopct='%1.2f%%', shadow=True)
plt.title('Pie Chart of Winners')
plt.axis('equal')
plt.legend(labels)
plt.show()
```

```python
# colors = ['#F4EA28', 'blue', 'red', 'green', 'orange']
colors = ['#F4EA28', 'orange']
# How is the output giving analysis ?
X = df['season']
Y = df['win_by_runs']
plt.bar(X, Y, color=colors)
plt.title('Bar Graph')
plt.show()
```
