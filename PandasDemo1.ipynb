```python
# Data Prepration/Cleaning
# dataset = pd.read_csv('data.csv')

# Check Missing Values
dataset.isnull().sum()
```

```python
dataset.head()
```

```python
# Drop the Columns with missing values
dataset = dataset.dropna(axis=1, how='all')
dataset.head()
```

```python
# Fill Columns with missing values
print(dataset['city'])
```

```python
dataset['city'].fillna('Bangalore', inplace=True)
```

```python
# Check Missing Values
dataset.isnull().sum()
```

```python
# Convert the column to the date data type
dataset['date'] = pd.to_datetime(dataset['date'])
print(dataset['city'][0])
```

```python
dataset = dataset.sort_values('date', ascending=False)
dataset = dataset.reset_index(drop=True)
dataset.head()
```

```python
dataset = pd.read_csv('data.csv')
dataset.head()
```

```python
# Scatter Plot Matrix
columns = ['id', 'season', 'dl_applied', 'win_by_runs', 'win_by_wickets']
pd.plotting.scatter_matrix(dataset[columns], figsize=(10,10))
plt.show()
```

```python
# Scatter Plot Matrix
columns = ['id', 'season', 'dl_applied', 'winner', 'win_by_runs', 'win_by_wickets']
pd.plotting.parallel_coordinates(dataset[columns], 'winner', colormap='rainbow')
plt.title('Parallel Plot')
plt.xlabel('Variables')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()
```
