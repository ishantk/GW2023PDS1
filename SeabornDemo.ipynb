```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.__version__
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
# Scatter Plot
sns.scatterplot(x='season', y='win_by_runs', data=df)
plt.title('Scatter Plot')
plt.show()
```

```python
# Line Plot
sns.lineplot(x='season', y='win_by_runs', data=df)
plt.title('Line Plot')
plt.show()
```

```python
# Histogram Plot
sns.histplot(x='win_by_runs', data=df)
plt.title('Histogram Plot')
plt.show()
```

```python
# Histogram Plot
sns.histplot(x='toss_decision', data=df)
plt.title('Histogram Plot')
plt.show()
```

```python
# Violin Plot
sns.set(style='whitegrid')
# sns.set(style='darkgrid')
colors = ['#F4EA28', 'blue', 'red', 'green', 'orange']
sns.violinplot(x='city', y='win_by_runs', data=df, palette=colors)
plt.title('Violin Plot')
plt.xticks(rotation=90)
plt.show()
```

```python
# KDE Plot
sns.kdeplot(x='win_by_runs', data=df)
plt.title('KDE Plot')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,8))

sns.set(style='whitegrid')
colors = ['#F4EA28', 'blue', 'red', 'green', 'orange']
sns.violinplot(x='city', y='win_by_runs', data=df, palette=colors, ax=ax1)
ax1.set_title('Violin Plot')

# KDE Plot
sns.kdeplot(x='win_by_runs', data=df, ax=ax2)
ax2.set_title('KDE Plot')

# SetUp All Graphs together
plt.tight_layout()
plt.show()
```
