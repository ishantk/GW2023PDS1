```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
X = df['season']
Y = df['win_by_runs']
Y1 = df['win_by_wickets']
```

```python
# HW: Explore Area Charts :)

# Create Combined Chart
fig, axes = plt.subplots()

# Bar Graph
axes.bar(X, Y, color='orange')
axes.set_xlabel('season')
axes.set_ylabel('win by runs')
axes.set_title('Bar Chart')

# Line Graph
axes_line = axes.twinx()
axes_line.plot(X, Y1, color='green')
axes_line.set_ylabel('win by wickets')

# SetUp Both Graphs together
plt.tight_layout()
plt.show()
```

```python
# Subplots
id_col = df['id']
season_col = df['season']
win_by_runs_col = df['win_by_runs']
win_by_wickets_col = df['win_by_wickets']
dl_applied_col = df['dl_applied']
```

```python
fix, axes = plt.subplots(2, 2, figsize=(20,20))

# Plot 1
axes[0, 0].plot(id_col, season_col, marker='o')
axes[0, 0].set_title('Seasons')

# Plot 2
axes[0, 1].scatter(id_col, win_by_runs_col, marker='o')
axes[0, 1].set_title('Win By Runs')

# Plot 3
axes[1, 0].bar(id_col, win_by_wickets_col)
axes[1, 0].set_title('Win By Wickets')

# Plot 4
axes[1, 1].plot(id_col, dl_applied_col, marker='o')
axes[1, 1].set_title('DL Applied')

# SetUp All Graphs together
plt.tight_layout()
plt.show()
```
